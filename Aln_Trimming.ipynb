```python
from Bio import AlignIO
import os

# ===== USER SETTINGS =====
input_dir = "/content/input"
alignment_file = "_out.260227230832401kKOVDE4A6t1T3mPn0Qx8vlsfnormal.fasta"      # file inside /content/input
output_file = "trimmed_alignment.fasta"

ref_start = 103   # start position in reference (ungapped)
ref_end = 813    # end position in reference (ungapped)
reference_index = 0  # which sequence to use as reference (0 = first)
# =========================

input_path = os.path.join(input_dir, alignment_file)
output_path = os.path.join(input_dir, output_file)

alignment = AlignIO.read(input_path, "fasta")
reference = alignment[reference_index]

ref_columns = []
ungapped_pos = 0

for col_idx, residue in enumerate(reference.seq):
    if residue != "-":
        ungapped_pos += 1
    if ref_start <= ungapped_pos <= ref_end:
        ref_columns.append(col_idx)

if not ref_columns:
    raise ValueError("Requested reference positions not found in alignment.")

start_col = ref_columns[0]
end_col = ref_columns[-1] + 1

trimmed_alignment = alignment[:, start_col:end_col]
AlignIO.write(trimmed_alignment, output_path, "fasta")

print(f"✅ Trimmed alignment saved to: {output_path}")
print(f"   Alignment columns kept: {start_col + 1}–{end_col}")
```
